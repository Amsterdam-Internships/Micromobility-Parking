### **Features**
The datasets contain different features. The most important/interesting features are the following:
- section_2018: The identification of the observed sidewalk section
- year: The year the observation was made
- day_part: The part of the day when the observation was made
- facility: The type of the observed parking facility (e.g. rack, staples, space)
- cap_bike: The bike parking capacity of the observed facility
- cap_moped: The moped parking capacity of the observed facility
- cap_brokenup: The parking capacity of the observed facility for a sidewalk segment with roadworks
- cap_private: The private parking capacity of the observed facility (e.g. a wall ring) 
- bike: The bicycle parking occupancy of the observed facility
- moped_yellow: The moped (yellow) parking occupancy of the observed facility
- moped_blue: The moped (blue) parking occupancy of the observed facility
- cargo_bike: The cargo bicycle parking occupancy of the observed facility
- motor: The motor parking occupancy of the observed facility
- four_wheeler: The four wheeler parking occupancy of the observed facility
- stock: The stock bicycle parking occupancy of the observed facility
- brokenup: The bicycle parking occupancy of the observed facility for a sidewalk segment with roadworks
- rent_shared_bike: The parking occupancy of shared bicycles for the observed facility

The 2018 and 2019 datasets have no features for the parking occupancy of shared vehicles (rent_shared_bike) and vehicles at locations with roadworks (brokenup). It is unclear whether the parking counts for those vehicles are included in other vehicle parking features or if they simply were not counted. 

To make the processing of the data easier, it was decided to add the missing attributes to the 2018 and 2019 data, with the parking counts set to 0.

```python
# Add missing columns to the data from 2018 and 2019
raw_2018['rent_shared_bike'] = 0
raw_2019['rent_shared_bike'] = 0
raw_2018['brokenup'] = 0
raw_2019['brokenup'] = 0
```

The values for the feature 'facility' differ between the 2018/2019 and 2020/2021 datasets. New features are added to each dataset to align the values.

```python
# Use substrings of the facility values for the 2020 and 2021 datasets to remove the numbers
data_2020['facility'] = data_2020['facility'].str[3:]
data_2021['facility'] = data_2021['facility'].str[3:]
```

```python
# Add features that describe the facility type and location
def add_facility(df):
    df['facility'] = df['facility'].str.rstrip()

    conditions = [
    df['facility'] == "Buiten voorziening",
    df['facility'] == "Rek",
    df['facility'] == "Nietjes",
    df['facility'] == "Vak",
    df['facility'] == "Bromfietsvak",
    df['facility'] == "Voor fietsenwinkel",
    df['facility'] == "Gemengd vak",
    df['facility'] == "Overig",
    df['facility'] == "Vak met nietjes"]
    choices_type = ['no_facility','rack', 'staples','space','moped_space','bike_shop', 'mixed_space','other','space_with_staples']
    choices_location = ['outside','inside', 'inside','inside','inside','inside', 'inside','inside','inside']
    df['facility_type'] = np.select(conditions, choices_type, default = None)
    df['facility_location'] = np.select(conditions, choices_location, default = None)

    return df
```

```python
data_2018 = add_facility(data_2018)
data_2019 = add_facility(data_2019)
data_2020 = add_facility(data_2020)
data_2021 = add_facility(data_2021)
```

### **Check for missing values**

```python
# Definition to check for missing values in the datasets
def check_for_missing():
    null_2018 = pd.Series(data_2018.isnull().sum(), name = '2018')
    null_2019 = pd.Series(data_2019.isnull().sum(), name = '2019')
    null_2020 = pd.Series(data_2020.isnull().sum(), name = '2020')
    null_2021 = pd.Series(data_2021.isnull().sum(), name = '2021')
    missing = pd.concat([null_2018, null_2019, null_2020, null_2021], axis=1)
    return missing
```

## **Change format of data from long to wide**

```python
# Create df with all datasets
dfs = [data_2018, data_2019, data_2020, data_2021]
data = pd.concat(dfs)
```

```python
# Create pivot table using the facility type
pivot_facility_type = pd.pivot_table(data, 
                       index=['section_2018','year','day_part'], 
                       columns=['facility_type'],
                       values=['cap_bike', 'cap_moped', 'cap_brokenup', 'cap_private', 
                               'bike', 'moped_yellow', 'moped_blue', 'cargo_bike', 'motor', 
                               'fourwheeler', 'rent_shared_bike','brokenup', 'stock', 
                               'capacity', 'occupancy'], 
                       aggfunc=np.sum,
                       fill_value=0)
pivot_facility_type.columns = pivot_facility_type.columns.to_flat_index().map(lambda x: '_in_'.join(x))
pivot_facility_type = pivot_facility_type.reset_index()
pivot_facility_type
```

```python
# Create pivot table using the facility location
pivot_facility_location = pd.pivot_table(data, index=['section_2018','year','day_part'], 
                        columns=['facility_location'], 
                        values=['cap_bike', 'cap_moped', 'cap_brokenup', 'cap_private', 
                                'bike', 'moped_yellow', 'moped_blue', 'cargo_bike', 'motor', 
                                'fourwheeler', 'rent_shared_bike', 'brokenup','stock', 
                                'capacity', 'occupancy'],
                        aggfunc=np.sum, 
                        fill_value=0)
pivot_facility_location.columns = pivot_facility_location.columns.to_flat_index().map(lambda x: '_'.join(x))
pivot_facility_location = pivot_facility_location.reset_index()
pivot_facility_location
```

```python
# Create df with the sum of parking capacity and occupancy per location per time
grouped_per_location = data.groupby(['section_2018','year','day_part']).sum().reset_index()
grouped_per_location
```

```python
# Merge all three dataframes together
merged = pivot_facility_type.merge(pivot_facility_location, on=['section_2018','year','day_part']).merge(grouped_per_location, on=['section_2018','year','day_part'])
merged
```

### **Add more features to the merged dataset**

```python
# Definition to add extra features
def add_features(df):
    df['perc_in'] = df['occupancy_inside'] / df['occupancy'] * 100
    df['perc_out'] = df['occupancy_outside'] / df['occupancy'] * 100
    df['net_point_occupancy'] = df['occupancy_inside'] / df['capacity'] * 100
    df['gross_point_occupancy'] = df['occupancy'] / df['capacity'] * 100
    
    df['perc_in'] = df['perc_in'].fillna(0)
    df['perc_out'] = df['perc_out'].fillna(0)
    df['net_point_occupancy'] = df['net_point_occupancy'].fillna(0)
    df['gross_point_occupancy'] = df['gross_point_occupancy'].fillna(0)

    df['gross_point_occupancy'].replace(np.inf, 250, inplace=True)
    df['net_point_occupancy'].replace(np.inf, 250, inplace=True)

    conditions = [
    (df['capacity'] == 0) & (df['occupancy_outside'] == 0),
    (df['capacity'] == 0) & (df['occupancy_outside'] > 0),
    (df['capacity'] > 0) & (df['occupancy_outside'] == 0),
    (df['capacity'] > 0) & (df['occupancy_inside'] < df['capacity']) & (df['occupancy_outside'] > 0),
    (df['capacity'] > 0) & (df['occupancy_inside'] == df['capacity']) & (df['occupancy_outside'] > 0),
    (df['capacity'] > 0) & (df['occupancy_inside'] > df['capacity']) & (df['occupancy_outside'] > 0),
    (df['capacity'] > 0) & (df['occupancy_inside'] < df['capacity']) & (df['occupancy_outside'] == 0),
    (df['capacity'] > 0) & (df['occupancy_inside'] == df['capacity']) & (df['occupancy_outside'] == 0)]
    choices_1 = ['No capacity, no occupancy','No capacity, but occupied','Capacity, but no occupancy, ','Capacity, but parked outside','Full capacity and parked outside','Overfull capacity and parked outside','Still capacity, no parked outside','Full capacity, no parked outside']
    choices_2 = [3,8,1,5,6,7,2,4]
    df['group_1'] = np.select(conditions, choices_1, default='?')
    df['group_2'] = np.select(conditions, choices_2, default='?')
    df['group_2'] = pd.to_numeric(df['group_2'])  

    return df
```
